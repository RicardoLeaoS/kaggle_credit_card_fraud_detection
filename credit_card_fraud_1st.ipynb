```python
def verifyKurtosisAndSkewness(values, bins=30):
    sns.displot(values, bins=bins)
    print('skewness ' + str(stats.skew(values)))
    print('kurtosis ' + str(stats.kurtosis(values)))
    return
```

## standardizing data

```python
y = new_df.iloc[:,-1]
X = new_df.iloc[:,:-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=22, stratify=y)
```

```python
std = preprocessing.StandardScaler()
```

```python
X_train = std.fit_transform(X_train)
```

```python
X_val = std.transform(X_test)
```

## Support Vector Classifier

```python
folds = 5
kfold = StratifiedKFold(n_splits=folds, shuffle=True)
mean_score = 0
```

```python
k = 1
for train_ix, val_ix in kfold.split(X_train, y_train):
    svc = SVC()
    # select rows
    train_X, val_X = X_train[train_ix], X_train[val_ix]
    train_y, val_y = y_train.iloc[train_ix], y_train.iloc[val_ix]
    
    svc.fit(train_X, train_y)
    y_val_pred = svc.predict(val_X)
    val_score =  average_precision_score(val_y, y_val_pred)
    mean_score += val_score / folds
    
    print(f"fold {k} : average_precision_score is : {val_score}")
    k+=1
        
print(f"mean average_precision_score is : {mean_score}")
```

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
svc_y_pred = svc.predict(X_test)
```

```python
average_precision_score(y_test, svc_y_pred)
```

```python
accuracy_score(y_test, svc_y_pred)
```

```python
cm = confusion_matrix(y_true=y_test, y_pred=svc_y_pred, normalize="true")

fig, ax = plt.subplots(figsize=(7,7))
sns.set(font_scale=2.0) # Adjust to fit
sns.heatmap(cm, annot=True, ax=ax, cmap="Blues", fmt=".2g")  

# Labels, title and ticks
label_font = {'size':'18'}  # Adjust to fit
ax.set_xlabel('Predicted labels', fontdict=label_font)
ax.set_ylabel('Observed labels', fontdict=label_font)

title_font = {'size':'21'}  # Adjust to fit
ax.set_title('Confusion Matrix', fontdict=title_font)

ax.tick_params(axis='both', which='major', labelsize=15)  # Adjust to fit
ax.xaxis.set_ticklabels(['False', 'True']);
ax.yaxis.set_ticklabels(['False', 'True']);
plt.show()

plt.tight_layout()

fig.savefig("svc_confusion_matrix.png")



```

## Random Forest Classifier

```python
folds = 5
kfold = StratifiedKFold(n_splits=folds, shuffle=True, random_state=11)
mean_score = 0
```

```python
k = 1
for train_ix, val_ix in kfold.split(X_train, y_train):
    rfc = RandomForestClassifier()
    # select rows
    train_X, val_X = X_train[train_ix], X_train[val_ix]
    train_y, val_y = y_train.iloc[train_ix], y_train.iloc[val_ix]
    
    rfc.fit(train_X, train_y)
    y_val_pred = rfc.predict(val_X)
    val_score =  average_precision_score(val_y, y_val_pred)
    mean_score += val_score / folds
    
    print(f"fold {k} : average_precision_score is : {val_score}")

    k+=1
    
print(f"mean average_precision_score is : {mean_score}")
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_y_pred = rfc.predict(X_test)
```

```python
average_precision_score(y_test, rfc_y_pred)
```

```python
accuracy_score(y_test, rfc_y_pred)
```

```python
roc_auc_score(y_test, rfc_y_pred)
```
